```python
import bs4
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://economictimes.indiatimes.com/markets/stocks/news"
```

```python
r = requests.get(url)
```

```python
soup = bs4.BeautifulSoup(r.text, 'html.parser')
```

```python
soup.find_all()
```

```python
soup.find_all('div')
```

```python
soup.find_all('div',{'class':'eachStory'})
```

```python
soup.find_all('div',{'class':'eachStory'})[0].find('span')
```

```python
for i in range (7):  
  print(f"\n{soup.find_all('div',{'class':'eachStory'})[i].find('h3').text}")
  print(f"{soup.find_all('div',{'class':'eachStory'})[i].find('p').text}")
  
```
