```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("forestfires.csv")
data
```

```python
monthly = data.groupby('month')['X'].sum()
monthly = monthly.sort_values(ascending=False)
monthly
```

```python
plt.bar(monthly.index,monthly.values)
plt.xlabel('Months')
plt.ylabel('Number of Fires')
```

```python
plt.boxplot([
    data[data['day']=='mon']['X'],
    data[data['day']=='tue']['X'],
    data[data['day']=='wed']['X'],
    data[data['day']=='thu']['X'],
    data[data['day']=='fri']['X'],
    data[data['day']=='sat']['X'],
    data[data['day']=='sun']['X'],
])
plt.xlabel('Day of week')
plt.ylabel('No of fires')

plt.show()
```

```python
plt.scatter(data['temp'],data['X'])
plt.xlabel('Temp')
plt.ylabel('No of fires')

plt.show()
```

```python
plt.hist(data['X'],bins=20)
plt.xlabel('No of Fires')
plt.ylabel('Frequency')

plt.show()
```

```python
corr = data.corr()
sns.heatmap(corr,annot=True)
plt.title('Correlation Matrix')
plt.show()
```
